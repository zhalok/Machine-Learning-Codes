```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
samples = 50
x = np.random.rand(samples)
y = 12*x + 100 + np.random.uniform(low=-1,high=1,size=50)
plt.scatter(x,y,s=5)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=.5)
```

```python

pf = PolynomialFeatures(degree=1000)
x_train_poly = pf.fit_transform(x_train.reshape(-1,1))
x_test_poly = pf.fit_transform(x_test.reshape(-1,1))

lin_reg = LinearRegression()
lin_reg.fit(x_train_poly,y_train)
training_acc = lin_reg.score(x_train_poly,y_train)
test_acc = lin_reg.score(x_test_poly,y_test)

```

```python
plt.title("Training set")
plt.scatter(x_train,y_train,c='r')
```

```python
plt.title("Testing Set")
plt.scatter(x_test,y_test,c='b')
```

```python
preds = lin_reg.predict(x_train_poly)
plt.scatter(x_train,y_train,c='b')
plt.plot(sorted(x_train),sorted(preds),c='r')
```

```python
preds = lin_reg.predict(x_test_poly)
plt.scatter(x_test,y_test,c='b')
plt.plot(sorted(x_test),sorted(preds),c='r')
```

```python
print("Training accuracy",training_acc)
print("Testing accuracy",test_acc)
```

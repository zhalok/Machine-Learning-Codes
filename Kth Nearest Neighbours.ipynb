```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import random
import numpy as np
import math

# Generate a classification dataset with 2 labels
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,
                           n_redundant=0, n_clusters_per_class=1,
                           random_state=42)
new_observe = np.random.random((1,2))

# Plot the generated dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.scatter(new_observe[:,0],new_observe[:,1],c='g')
# print(new_observe)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Classification Dataset')
plt.show()
```

```python
def euclid_distance(p1,p2):
    x1,y1 = p1
    x2,y2 = p2
    
    return math.sqrt( (x1-x2)**2 + (y1-y2)**2  )
    
```

```python
def KNN(new_observe,X,y,k):
    
    distances = []
    
    for index,value in enumerate(X):
        distance = euclid_distance(value,new_observe.flatten())
        distances.append([distance,y[index]])

    distances.sort(key=lambda x: x[0])
    
#     k=50
    distances = np.array(distances)
    print("distances",len(distances))
    print("first k neighbour distances",(distances[0:k]))
    labels = distances[0:k][:,1]

    zeros = 0
    ones = 0
    
    for i in labels:
        if i == 0:
            zeros += 1
        else:
            ones += 1

    
    if zeros>ones:
        return 0
    else:
        return 1
```

```python
predict = KNN(new_observe,X,y,50)
predict
```

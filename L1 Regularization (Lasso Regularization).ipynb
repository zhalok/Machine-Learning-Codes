```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
import pandas as pd
```

```python
samples = 50
x = np.random.rand(samples)
y = 12*x + 100 + np.random.uniform(low=-1,high=1,size=50)
plt.scatter(x,y,s=5)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x.reshape(-1,1),y)
preds = lin_reg.predict(x.reshape(-1,1))

plt.scatter(x,y,c='b',s=5)
plt.scatter(x,preds,c='g',s=6)
```

```python
plf = PolynomialFeatures(degree=1000)
x_poly = plf.fit_transform(x.reshape(-1,1))

poly_reg = LinearRegression()
poly_reg.fit(x_poly,y)
preds = poly_reg.predict(x_poly)

plt.scatter(x,y,c='b',s=5)
plt.scatter(x,preds,c='r',s=8)
```

```python
lasso_reg = Lasso(alpha=1,max_iter=5000)
lasso_reg.fit(x_poly,y)
preds = lasso_reg.predict(x_poly)

plt.scatter(x,y,c='b',s=5)
plt.scatter(x,preds,c='r',s=8)

acc = lasso_reg.score(x_poly,y)
acc
```

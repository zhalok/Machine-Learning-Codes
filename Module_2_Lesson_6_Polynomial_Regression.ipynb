```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
#lets create a random data set which is not linear

x = 2 - 3 * np.random.normal(0, 1, 20)

y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)
print(x.shape)
print(y.shape)
plt.scatter(x,y)
```

```python
x_ = x.reshape(-1,1)
print(x_.shape)
```

```python
linear_reg_model = LinearRegression()
linear_reg_model.fit(x_,y)
```

```python
plt.plot(x,linear_reg_model.predict(x_))
plt.scatter(x,y)
```

```python
polynomial_features = PolynomialFeatures(degree=3)
x_poly = polynomial_features.fit_transform(x_)
print(x_poly)
```

```python
linear_reg_model.fit(x_poly,y)
```

```python
preds = linear_reg_model.predict(x_poly)
plt.plot(sorted(x),sorted(preds))
plt.scatter(x,y)
```

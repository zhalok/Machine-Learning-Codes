```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import KernelPCA
from mpl_toolkits import mplot3d
from sklearn.svm import SVC
```

```python
def plot(x,y,title="plotting"):
    
    plt.scatter(x[:, 0], x[:, 1], c=y, cmap='viridis')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
    plt.show()
    
```

```python
np.random.seed(42)


radius_inner = 2.0
radius_outer = 4.0
theta = np.linspace(0, 2 * np.pi, 500)
X_inner = np.array([np.cos(theta) * radius_inner, np.sin(theta) * radius_inner]).T
X_outer = np.array([np.cos(theta) * radius_outer, np.sin(theta) * radius_outer]).T


noise_inner = np.random.normal(scale=0.3, size=X_inner.shape)
noise_outer = np.random.normal(scale=0.3, size=X_outer.shape)
X_inner += noise_inner
X_outer += noise_outer

X = np.concatenate((X_inner, X_outer))
y = np.concatenate((np.zeros(len(X_inner)), np.ones(len(X_outer))))


plot(X,y)
```

```python



svm = SVC(kernel='rbf',C=10)
svm.fit(X, y)
accuracy = svm.score(X,y)
print("Accuracy",accuracy)

plot3d(X,y,svm)
```

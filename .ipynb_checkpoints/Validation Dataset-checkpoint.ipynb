```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
#lets create a random data set which is not linear

x = 2 - 3 * np.random.normal(0, 1, 20)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)
print(x.shape)
print(y.shape)
plt.scatter(x,y)
```

```python
reg = DecisionTreeRegressor(max_depth=5)
reg.fit(x.reshape(-1,1),y)
preds = reg.predict(x.reshape(-1,1))

plt.scatter(x,y)
plt.plot(sorted(x),sorted(preds))
```

```python
legends = []
plots = []
train_accs = []
testin_accs = []

for depth in range(1,1000):
    reg = DecisionTreeRegressor(max_depth=depth)
    x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=.5)
    reg.fit(x_train.reshape(-1,1),y_train)
    preds = reg.predict(x.reshape(-1,1))

    plots.append([sorted(x),sorted(preds)])

    training_acc = reg.score(x_train.reshape(-1,1),y_train)
    testing_acc = reg.score(x_test.reshape(-1,1),y_test)
    legends.append("Max Depth "+str(depth))

#     print("training_acc",training_acc)
#     print("testing_acc",testing_acc)
    
    train_accs.append(training_acc)
    testin_accs.append(testing_acc)
    

for index,plot in enumerate(plots):
    plt.plot(sorted(plot[0]),sorted(plot[1]),c="C"+str(index))
# plt.legend(legends)
```

```python
x = x.reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.5)
x_train, x_val, y_train, y_val = train_test_split(x_train,y_train,train_size=0.5)

train_accs = []
val_accs = []

for depth in range(1,1000):
    reg = DecisionTreeRegressor(max_depth=depth)
    reg.fit(x_train,y_train)
    train_acc = reg.score(x_train,y_train)
    val_acc = reg.score(x_val,y_val)
    train_accs.append(train_acc)
    val_accs.append(val_acc)

rng = range(1,1000)
plt.plot(sorted(rng),sorted(train_accs)) 
plt.plot(sorted(rng),sorted(val_accs))
```

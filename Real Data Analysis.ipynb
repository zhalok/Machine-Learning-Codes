```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
from sklearn.preprocessing import LabelEncoder
import seaborn as sb
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("titanic/train.csv")
dataset = dataset.drop(["Name","Ticket","PassengerId","Fare"],axis=1)
# le = LabelEncoder()

dataset.head()
# print(len(dataset))
```

```python
le = LabelEncoder()
dataset["Sex"] = le.fit_transform(dataset["Sex"])
dataset["Cabin"] = le.fit_transform(dataset["Cabin"])
dataset["Embarked"] = le.fit_transform(dataset["Embarked"])
dataset["Cabin"] = le.fit_transform(dataset["Cabin"])

dataset_ = dataset["Age"].dropna()

mean_age = np.mean(dataset_.values)

dataset = dataset.fillna(mean_age)

# dataset.isna().sum()

# age = dataset["Age"].tolist()
# for a in age:
# #     print(a)
#     if a == 'nan':
#         print("hello")


# for i in range(len(dataset)):
#     print(dataset.iloc[i]["Age"])
```

```python
x = dataset.drop("Survived",axis=1)
y = dataset["Survived"]

train_x, test_x, train_y, test_y = train_test_split(x,y,train_size=0.2)
clf = DecisionTreeClassifier()
clf.fit(train_x,train_y)
clf.score(test_x,test_y)
```
